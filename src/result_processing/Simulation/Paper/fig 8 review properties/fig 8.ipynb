```python
import numpy as np
import matplotlib.pyplot as plt

# Definir los puntos iniciales y finales
a = (31, 108)
b = (32, 109)
c = (32, 109)
d = (25, 90)

# Número de valores a interpolar
num_values = 11

# Interpolación lineal para cada componente
a_values = np.linspace(a[0], a[1], num_values)
b_values = np.linspace(b[0], b[1], num_values)
c_values = np.linspace(c[0], c[1], num_values)
d_values = np.linspace(d[0], d[1], num_values)

# Crear la figura y el eje
fig, ax = plt.subplots(figsize=(10, 7))

# Definir los valores del eje x para que correspondan al rango de 0 a 100%
x_values = np.linspace(0, 100, num_values)

# Graficar cada columna de valores interpolados con el nuevo rango del eje x
ax.plot(x_values, a_values, 's-', label='a', color = 'blue')
ax.plot(x_values, b_values, 'o-', label='b', color = 'r')
ax.plot(x_values, c_values, 'x-', label='c', color='black')
ax.plot(x_values, d_values, '^-', label='d', color = 'green')

# Añadir leyenda
ax.legend()

# Añadir etiquetas y título
ax.set_xlabel("SOC (%)")  # Se ajusta el eje x a porcentaje
ax.set_ylabel("Elastic Modulus")
ax.set_title("Anode")
# Mostrar el gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Times New Roman"
# Definir los puntos iniciales y finales para las series
a = (31, 108)
b = (32, 109)
c = (32, 109)
d = (25, 90)
c1 = (1.954, 1.934)
d1 = (2.279, 2.18)
b2 = (264, 59.8)
c2 = (178, 89)
d2 = (230, 145)
e2 = (178, 100)
c3 = (3.4, 3.4)
d3 = (5.02, 4.8)
e3 = (4.8, 4.8)

# Número de valores a interpolar
num_values = 11

# Interpolación lineal para cada componente
a_values = np.linspace(a[0], a[1], num_values)
b_values = np.linspace(b[0], b[1], num_values)
c_values = np.linspace(c[0], c[1], num_values)
d_values = np.linspace(d[0], d[1], num_values)
c1_values = np.linspace(c1[0], c1[1], num_values)
d1_values = np.linspace(d1[0], d1[1], num_values)
b2_values = np.linspace(b2[0], b2[1], num_values)
c2_values = np.linspace(c2[0], c2[1], num_values)
d2_values = np.linspace(d2[0], d2[1], num_values)
e2_values = np.linspace(e2[0], e2[1], num_values)
c3_values = np.linspace(c3[0], c3[1], num_values)
d3_values = np.linspace(d3[0], d3[1], num_values)
e3_values = np.linspace(e3[0], e3[1], num_values)

# Valores del eje x
x_values = np.linspace(0, 100, num_values)

# Crear la figura y subplots 2x2
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# Gráficas Anode
ax[0, 0].plot(x_values, a_values, 's-', label='a', color='blue')
ax[0, 0].plot(x_values, b_values, 'o-', label='b', color='red')
ax[0, 0].plot(x_values, c_values, 'x-', label='c', color='black')
ax[0, 0].plot(x_values, d_values, '^-', label='d', color='green')
ax[0, 0].set_title('Anode Elastic Modulus')
ax[0, 0].set_xlabel('SOC (%)')
ax[0, 0].set_ylabel('Elastic Modulus')
ax[0, 0].legend()

# Gráficas Anode Density
ax[0, 1].plot(x_values, c1_values, 'x-', label='c', color='black')
ax[0, 1].plot(x_values, d1_values, '^-', label='d', color='green')
ax[0, 1].set_title('Anode Density')
ax[0, 1].set_xlabel('SOC (%)')
ax[0, 1].set_ylabel('Density')
ax[0, 1].legend()

# Gráficas Cathode Elastic Modulus
ax[1, 0].plot(x_values, b2_values, 'o-', label='b', color='red')
ax[1, 0].plot(x_values, c2_values, 'x-', label='c', color='black')
ax[1, 0].plot(x_values, d2_values, '^-', label='d', color='green')
ax[1, 0].plot(x_values, e2_values, '^-', label='e', color='orange')
ax[1, 0].set_title('Cathode Elastic Modulus')
ax[1, 0].set_xlabel('SOC (%)')
ax[1, 0].set_ylabel('Elastic Modulus')
ax[1, 0].legend()

# Gráficas Cathode Density
ax[1, 1].plot(x_values, c3_values, 'x-', label='c', color='black')
ax[1, 1].plot(x_values, d3_values, '^-', label='d', color='green')
ax[1, 1].plot(x_values, e3_values, '^-', label='e', color='orange')
ax[1, 1].set_title('Cathode Density')
ax[1, 1].set_xlabel('SOC (%)')
ax[1, 1].set_ylabel('Density')
ax[1, 1].legend()

# Ajustar el layout y mostrar el gráfico
plt.tight_layout()
plt.show()
```
