```python
import os
import numpy as np
import matplotlib.pyplot as plt

def plot_npy_files(directory):
    # Get all .npy files in the directory
    npy_files = [f for f in os.listdir(directory) if f.endswith('.npy')]
    npy_files.sort()  # Sort files alphabetically (optional)
    
    num_files = len(npy_files)
    num_subplots = min(num_files, 11)  # Maximum 11 plots
    
    if num_files == 0:
        print("No .npy files found in the directory.")
        return
    
    fig, ax = plt.subplots(figsize=(20, 15))  # Adjusted for a tall layout
    
    soc_labels =(2, 4, 8, 16, 64, 128, 256)  # Generate SOC labels from 0% to 100%
    offset = 1.5  # Vertical offset between lines   
    
    for i, file in enumerate(npy_files[:11]):
        file_path = os.path.join(directory, file)
        data = np.load(file_path)
        
        if data.ndim < 2:
            print(f"Skipping {file}: Not a 2D array.")
            continue
        
        last_column = data[-1,:]  # Extract last column
        x_values = np.arange(len(last_column)) * 5e-4  # Adjust x-axis scale
        
        ax.plot(x_values, last_column + i * offset, color='black')  # Offset each line
        
        # Add text label to the right side
        ax.text(x_values[-1] + 0.2 * (x_values[-1] - x_values[0]), 
                last_column[-1] + i * offset, 
                f"{int(soc_labels[i])}% SOC", 
                verticalalignment='center', fontsize=15)
    
    ax.set_xlabel(("Time "u"\u03bcs"), fontsize=20)
    ax.set_ylabel("Amplitude", fontsize=20)
    ax.set_yticks([])  # Remove y-ticks
    ax.spines['left'].set_visible(True)  # Hide left spine
    ax.spines['right'].set_visible(False)  # Hide right spine
    
    plt.tight_layout()
    plt.show()

    # Example usage
plot_npy_files("./")
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

# Ruta de los archivos .npy
url = './'

# Pedir al usuario el nombre del archivo
archivo = 'nodes_f600_dt_f0.001_layer_n_2.npy'

# Construir la ruta del archivo
ruta = os.path.join(url, archivo)

# Verificar si el archivo existe
if os.path.isfile(ruta) and archivo.endswith('.npy'):
    # Cargar los datos del archivo .npy
    data = np.load(ruta)
    ultimo_nodo = data[-1, 0:1750]

    # Obtener el valor máximo y su posición
    valor = ultimo_nodo.max()
    print(f"Valor máximo: {valor}")
    min_valor = ultimo_nodo.min()
    print(f"Valor mínimo: {min_valor}")

    # Generar el tiempo en el eje x
    tiempo = np.arange(len(ultimo_nodo))

    # Crear la figura y el subplot
    fig, ax = plt.subplots(figsize=(12, 5))

    # Graficar los datos
    ax.plot(tiempo, ultimo_nodo, color='black', label=f'Tiempo = 1.6 us')  # tiempo en el eje x
    ax.set_title('Transmitted and Reflected Amplitude (Cathode - Anode)', fontsize=16)
    ax.set_xlabel('Length (mm)', fontsize=14)
    ax.set_ylabel('Amplitude', fontsize=14)
    ax.legend(loc='best', fontsize=12)

    # Ajustar diseño y mostrar gráfico
    plt.tight_layout()
    plt.show()

else:
    print("El archivo no existe o no es un archivo .npy válido. Verifique el nombre e intente nuevamente.")
```
